```python
import os
import re

def modify_files_in_directory(directory_path):
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.endswith('.cs'):
                file_path = os.path.join(root, file)
                print(f'Checking: {file_path}')
                modify_file_if_needed(file_path)

def modify_file_if_needed(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        
    # Check if the file contains 'PathSpawnerModel'
    if 'PathSpawnerModel' in content:
        print(f'Found: {file_path}')
        # Regular expression to find the pattern and add an empty string parameter
        modified_content = re.sub(r'(new SplitterModel\("[^"]*", new string\[\]\s*\{[^}]*\})\)', r'\1, "")', content)
        
        # Only write back if changes were made
        if modified_content != content:
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(modified_content)
            print(f'Modified: {file_path}')

modify_files_in_directory("./")
```
